```julia
function calc_residuals(y, μ, ϕ, θ)
    ϵ = [μ-y[1]]
    y_hat = [μ]
    T = size(y, 1)
    for t in 2:T
        yt_hat = μ+ϕ*(y[t-1]-μ)+θ*ϵ[t-1]
        ϵt = y[t]-yt_hat
        push!(ϵ, ϵt)
        push!(y_hat, yt_hat)
    end
    return ϵ
end
```

```julia
normal_residuals_S = calc_residuals(P_t_S, μ_S, ϕ_S, θ_S)
```

```julia
D = MvNormal([mean(normal_residuals_S),mean(normal_residuals_W)],Cov_mat)
```

```julia
test = rand(D)
```

```julia
function Residual_Sample(Dist, μ_S_res, μ_W_res, T, kmT)
    y1 = []
    y2 = []
    ykm_1= []
    ykm_2= []
    for t in 1:T
        yt = rand(Dist)
        push!(y1, yt[1])
        push!(y2, yt[2])
    end
    y = [y1 y2]
    dist = pairwise(Euclidean(), y, dims=1)
    kmed = kmedoids(dist,kmT)
    for i in 1:kmT
        push!(ykm_1, μ_S_res + y1[kmed.medoids[i]])
        push!(ykm_2, μ_W_res + y2[kmed.medoids[i]])
    end
    y = [ykm_1 ykm_2]
    return y
end
```

```julia
Residuals = Residual_Sample(D, mean(normal_residuals_S), mean(normal_residuals_W), 100, 4)
```

```julia
price_scenarios =
[50 40 1 1.5
50 53 1 1.5
20 10 1 1.5
20 23 1 1.5
50 40 0.9 1
50 53 0.9 1
20 10 0.9 1
20 23 0.9 1]
```

```julia
function sample_PtauPt(Residual_Sample, prices, ϵ_Dist, μ_S, ϕ_S, θ_S, μ_W, ϕ_W, θ_W, T, kmT)
    PT_S= []
    PT_W = []
    results = []
    
    P_tau = []
    P_tau_S = []
    P_tau_W = []
    P_t = []
    P_t_S = []
    P_t_W = []
    lambda_D = []
    lambda_A = []
    r_plus = []
    r_minus = []
    
    #Add Header
    push!(P_tau,"P_tau")
    push!(P_tau_S,"P_tau_S")
    push!(P_tau_W,"P_tau_W")
    push!(P_t,"P_t")
    push!(P_t_S,"P_t_S")
    push!(P_t_W,"P_t_W")
    push!(lambda_D,"lambda_D")
    push!(lambda_A,"lambda_A")
    push!(r_plus,"r_plus")
    push!(r_minus,"r_minus")
    
    for i in 1:size(Residual_Sample, 1)
        yt_S = []
        yt_W = []
        ykm_S = []
        ykm_W = []
        for t in 1:T
            ϵt = rand(ϵ_Dist)
            #y0 = μ
            #y1 = μ + ϵt
            #y2 = μ+ϕ*(y[t-1]-μ)+ϵt+θ*ϵ[t-1]
            yt_S = μ_S+ϕ_S*(Residual_Sample[i,1])+ϵt[1]+θ_S*(Residual_Sample[i,1])
            yt_W = μ_W+ϕ_W*(Residual_Sample[i,2])+ϵt[2]+θ_W*(Residual_Sample[i,2])
            push!(PT_S, yt_S)
            push!(PT_W, yt_W)
        end
        y = [PT_S PT_W]
        dist = pairwise(Euclidean(), y, dims=1)
        kmed = kmedoids(dist,kmT)
        for j in 1:kmT
            push!(ykm_S, PT_S[kmed.medoids[j]])
            push!(ykm_W, PT_W[kmed.medoids[j]])
        end
        
        for a in 1:size(prices,1)
            for b in 1:kmT
                
                push!(P_tau,μ_S + Residual_Sample[i,1] + μ_W + Residual_Sample[i,2])
                push!(P_tau_S,μ_S + Residual_Sample[i,1])
                push!(P_tau_W,μ_W + Residual_Sample[i,2])
                push!(P_t,ykm_S[b] + ykm_W[b])
                push!(P_t_S,ykm_S[b])
                push!(P_t_W,ykm_W[b])
                push!(lambda_D,price_scenarios[a,1])
                push!(lambda_A,price_scenarios[a,2])
                push!(r_plus,price_scenarios[a,3])
                push!(r_minus,price_scenarios[a,4])
            end
        end
    end
    results = [P_tau P_tau_S P_tau_W P_t P_t_S P_t_W lambda_D lambda_A r_plus r_minus]
    return results
end
```

```julia
Results = sample_PtauPt(Residuals, price_scenarios, D, μ_S, ϕ_S, θ_S, μ_W, ϕ_W, θ_W, 100, 4)
```
